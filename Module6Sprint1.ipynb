```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.read_csv('data.csv')
data.plot(x='Cost', y='Profit', kind='scatter')

plt.show()
```

```python
plt.scatter(x='Price', y='Cost', data=data)
plt.title('Price vs Cost scatter')
plt.xlabel('Price')
plt.ylabel('Cost')
plt.show()
```

```python
plt.scatter(x=data['Type']=='Juice', y='Name', data=data)

plt.show()

# Method 2

plt.scatter(x=data['Profit'], y=data['VAT returned'])
sns.regplot(data['Profit'],data['Cost'])
plt.show()
#The first method uses the argument data which specifies the data source, whereas the second method directly uses dataframe slicing and hence, there is no need to
#specify the data argument.

#now we are going to go throug histogram plots 
# -Example 23-

# Data values for creating a histogram

y = data['Profit']

# Creating a histogram

plt.hist(y)

plt.xlabel('Profit')

plt.ylabel('Frequency')

plt.show()

# -Example 24-

# Data values for creating a histogram

y = data['Cost']
# Creating a histogram

plt.hist(y, bins= 20)

plt.xlabel('Cost')

plt.ylabel('Frequency')

plt.show()

# -Example 25-

# Creating an array

array = np.random.normal(0, 1.5, 10000)

# Creating a histogram
plt.hist(array)

plt.xlabel('Bins')

plt.ylabel('Frequency')

plt.show()

data.describe()
```
